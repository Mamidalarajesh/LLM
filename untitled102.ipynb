```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity


nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')


covid_metadata = pd.read_excel('COVID.metadata.xlsx')
lung_opacity_metadata = pd.read_excel('Lung_Opacity.metadata.xlsx')
normal_metadata = pd.read_excel('Normal.metadata.xlsx')
viral_pneumonia_metadata = pd.read_excel('Viral Pneumonia.metadata.xlsx')


print("COVID Metadata Columns:", covid_metadata.columns.tolist())
print("Lung Opacity Metadata Columns:", lung_opacity_metadata.columns.tolist())
print("Normal Metadata Columns:", normal_metadata.columns.tolist())
print("Viral Pneumonia Metadata Columns:", viral_pneumonia_metadata.columns.tolist())


relevant_column = 'Impression'


if relevant_column in covid_metadata.columns:

    print("Missing values in the relevant column:", covid_metadata[relevant_column].isnull().sum())
    covid_metadata[relevant_column].fillna('', inplace=True)  # Fill NaN with an empty string


    def preprocess_text(text):
        text = text.lower()
        words = nltk.word_tokenize(text)
        words = [word for word in words if word.isalpha() and word not in stopwords.words('english')]
        lemmatizer = WordNetLemmatizer()
        words = [lemmatizer.lemmatize(word) for word in words]
        return ' '.join(words)


    covid_metadata['Processed_Impression'] = covid_metadata[relevant_column].apply(preprocess_text)
    lung_opacity_metadata['Processed_Impression'] = lung_opacity_metadata[relevant_column].apply(preprocess_text)
    normal_metadata['Processed_Impression'] = normal_metadata[relevant_column].apply(preprocess_text)
    viral_pneumonia_metadata['Processed_Impression'] = viral_pneumonia_metadata[relevant_column].apply(preprocess_text)


    all_reports = pd.concat([covid_metadata, lung_opacity_metadata, normal_metadata, viral_pneumonia_metadata])


    vectorizer = CountVectorizer()
    X = vectorizer.fit_transform(all_reports['Processed_Impression'])


    cosine_sim = cosine_similarity(X)


    top_pairs = []
    for i in range(len(cosine_sim)):
        similar_indices = cosine_sim[i].argsort()[-2:-12:-1]
        for index in similar_indices:
            if index != i:
                top_pairs.append((all_reports.iloc[i]['Processed_Impression'], all_reports.iloc[index]['Processed_Impression']))


    top_pairs_df = pd.DataFrame(top_pairs, columns=['Word Pair 1', 'Word Pair 2'])


    plt.figure(figsize=(10, 8))
    plt.barh(top_pairs_df['Word Pair 1'].head(10), range(10), color='blue')
    plt.xlabel('Index')
    plt.title('Top 10 Word Pairs')
    plt.show()

else:
    print(f"The column '{relevant_column}' does not exist in the DataFrame.")
```
